```python
import numpy as np
import numpy.linalg as la
```

```python
W = [[0.58, 0.24],
     [0.24, 0.72]]

(l1, l2), q = la.eigh(W)
s = np.array([[l1,  0],
              [ 0, l2]])
print(f'Q={q}\nΛ={s}\nQ^-1={q.T}')  
# these are the correct values of Q and Q^-1 (they are wrong in exercise sheet)
```

```python
q @ s @ q.T
```
